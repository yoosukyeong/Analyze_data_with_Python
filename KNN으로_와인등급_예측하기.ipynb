```python
X = data.drop('class', axis = 1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
mm_scaler = MinMaxScaler()
```

```python
mm_scaler.fit(X_train)
```

```python
scaled_X_train = mm_scaler.transform(X_train)
scaled_X_test = mm_scaler.transform(X_test)
```

```python
scaled_X_test[:3]
```

```python
mm_scaler = MinMaxScaler()
```

```python
scaled_X_train = mm_scaler.fit_transform(X_train)
```

```python
mm_scaler.transform(X_test)[:3]
```

```python
scaled_X_test = mm_scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(scaled_X_train, y_train)
```

```python
pred = knn.predict(scaled_X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(scaled_X_train, y_train)
pred = knn.predict(scaled_X_test)
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(scaled_X_train, y_train)
pred = knn.predict(scaled_X_test)
accuracy_score(y_test, pred)
```

```python
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(scaled_X_train, y_train)
    pred = knn.predict(scaled_X_test)
    print(i, accuracy_score(y_test, pred))
```

```python
accuracy_score_list = []
```

```python
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(scaled_X_train, y_train)
    pred = knn.predict(scaled_X_test)
    accuracy_score_list.append(accuracy_score(y_test, pred))
```

```python
sns.lineplot(x = range(1, 21), y = accuracy_score_list)
plt.xticks(range(1, 21))
plt.show()
```
